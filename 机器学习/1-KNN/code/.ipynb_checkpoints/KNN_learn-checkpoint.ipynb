```python
import numpy as np
```

```python
x = np.array([[172,60,42],
             [168,56,40],
             [189,70,44],
             [160,45,36],
             [163,50,42],
             [170,65,41]])
#给上面的特征标签加数据
y = np.array(['男','女','男','女','男','女'])
```

```python
#构造一组未知数据
x_test = np.array([[175,75,43],[170,60,38],[165,50,40]])
```

```python
#导入相关算法
from sklearn.neighbors import KNeighborsClassifier
#sklearn 全称scikit-learn，它是一个机器学习的系统库，几乎包含了所有机器学习的经典算法
#另外这个库还包含了大量的实验数据，还有一些数据清洗、处理的算法
#sklearn这个库封装度极高，对于开发者基本透明
```

```python
#创建算法模型，k-近邻分类模型
knn = KNeighborsClassifier(n_neighbors=3)
#n_neighors代表knn取前k个邻居，默认为5，一般不超过20
```

```python
#用创建好的算法模型，对已有的数据进行训练
knn.fit(x,y)
#在k-近邻里面，训练过程就是告诉knn对象，x里面的数据对应到y中代表什么
```

```python
#根据已有的数据基础对未知的数据进行预测（上面knn已经训练过了）
knn.predict(x_test)
#分别拿x_test中的每一个数据和已知的数据进行k近邻运算
```

### 用于分类

```python
#导入sklearn的官方数据
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
#提取特征数据
data= iris.data
#提取标签数据
target = iris.target
data.shape,target.shape
```

```python
#【注意】机器学习最后需要进行性能评测，需要把我们的数据提取出来一部分作为测试数据
#sklearn提供了一个训练数据和测试数据的切分方法
from sklearn.model_selection import train_test_split
```

```python
#这个方法首先会把data和target随机打乱，提取10%数据（包括特征和标签）作为测试数据
x_train,x_test,y_train,y_test = train_test_split(data,target,test_size=0.1)
#这个函数返回一个列表，包含训练特征、测试特征、训练标签、测试标签
```

```python
#创建你模型
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
#预测
y_ = knn.predict(x_test)
```

```python
y_,y_test
```

```python
#性能评测，查看准确率
knn.score(x_test,y_test)
```
